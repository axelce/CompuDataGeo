```julia
using Plots
using CSV
using GeoStats
using DataFrames
using LaTeXStrings
```

```julia
ALT03 = CSV.read("2018-03-altz_hora_L1.csv");
QRO03 = CSV.read("2018-03-jqro_hora_L1.csv");
MIN03 = CSV.read("2018-03-mine_hora_L1.csv");
MER03 = CSV.read("2018-03-meda_hora_L1.csv");
MZT03 = CSV.read("2018-03-mazt_hora_L1.csv");
AGS03 = CSV.read("2018-03-agsc_hora_L1.csv");
```

AGUASCALIENTES; REGIÓN CENTRO-OCCIDENTE

```julia
#División de datos, para un mes (marzo, 2018; AGS), por categoría. 
AGS03date = AGS03[:,1];
AGS03time = AGS03[:,2];
AGS03temp = AGS03[:,3];
AGS03hum = AGS03[:,4];
AGS03WSA = AGS03[:,5];
AGS03WSM = AGS03[:,6];
AGS03WD = AGS03[:,7];
AGS03PRES = AGS03[:,8];
```

```julia
#Histogramas para Temperatura, Humedad, Velocidad del viento promedio y Presión para un mes. 
pyplot()
h1 = histogram(AGS03temp,bins=20, title="Temperatura", ylabel=L"Temperatura ($°C$)", ylims=(0,95),legend=false,
    color="darkorange", alpha=0.7)
h2 = histogram(AGS03hum, bins=20,title="Humedad",ylabel="Humedad (%)", ylims=(0,120), legend=false,
    color="darkorange", alpha=0.7)
h3 = histogram(AGS03WSA, title="Vel. Prom. del Viento",ylabel=L"Vel. viento ($m/s^{2}$)", ylims=(0,165), legend=false,
    color="darkorange", alpha=0.7)
h4 = histogram(AGS03PRES, title="Presión",ylabel=L"Presión ($hPa/m^{2}$)", ylims=(0,149), legend=false,
    color="darkorange", alpha=0.7)



plot(h1,h2,h3,h4,layout=(2,2),box=:true,
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1, box=:true)
```

```julia
#Temperatura
println("Temperatura promedio ="," ",mean(AGS03temp), ", ", "Temperatura máxima =", " ", maximum(AGS03temp),
", ", "Temperatura mínima =", " ", minimum(AGS03temp))

#Humedad
println("Humedad promedio ="," ",mean(AGS03hum), ", ", "Humedad máxima =", " ", maximum(AGS03hum),
", ", "Humedad mínima =", " ", minimum(AGS03hum))

#Vel. Prom. del viento
println("Vel. del viento promedio ="," ",mean(AGS03WSA), ", ", "Vel. del viento máxima =", " ", maximum(AGS03WSA),
", ", "Vel. del viento mínima =", " ", minimum(AGS03WSA))

#Presión
println("Presión promedio ="," ",mean(AGS03PRES), ", ", "Presión máxima =", " ", maximum(AGS03PRES),
", ", "Presión mínima =", " ", minimum(AGS03PRES))
```
