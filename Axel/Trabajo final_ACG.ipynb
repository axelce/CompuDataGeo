<h3>INTRODUCCIÓN</h3>

<br />
**I. ATMÓSFERA**
<br />
<br />
La **atmósfera terrestre** es la capa gaseosa superficial que cubre a la *corteza* [*terrestre*]. Se le considera una mezcla mecánica de gases, compuesta, principalmente, por Nitrógeno (78.08 %), Oxígeno (20.95 %), Argón (0.93 %), Dióxido de Carbono (0.035 %), entre otros (0.005%). Los gases de Nitrógeno ($N_{2}$), Oxígeno ($O_{2}$)  y Argón ($Ar$) están mezclados en una proporción notablemente constante hasta unos 80 km de altura (Barry & Chorley, 1999). La atmósfera puede dividirse, convencionalmente, en varias capas horizontales, si se toma como base la temperatura (*Figura 1*). 
<br />
<br />
<img style="float: center;" src="atm.jpg" width="500">
<center>*Figura 1.* Las capas de la atmósfera. Obtenida de [http://www.eltiempoenchivilcoy.com.ar/las-capas-de-la-atmosfera/].</center>
<br />
<br />
**II. TEMPERATURA ATMOSFÉRICA**
<br />
<br />
El  aire  de  la  troposfera  se  calienta  a  partir del  calor  emitido  por  la  superficie  terrestre.  La  
temperatura  es  máxima  en  la  superficie  terrestre, alrededor  de  15  ºC  de  media,  y  a  partir  de  ahí  
comienza  a  descender  con  la  altura según  un  *Gradiente  Térmico  Vertical* (*GTV*)  de  6.5  °C  de  
descenso cada Km que se asciende en altura (la temperatura baja 0,65 °C cada 100m de altura) hasta llegar a -70 °C a los 12 km de altura. A partir de aquí, la temperatura asciende con la altura hasta llegar  próximo  a  los  0  °C  en  los  50  km.  Este  incremento  de  temperatura  está  relacionado  con  la absorción  por  el  ozono  de  la  radiación  solar  ultravioleta.  De  los  50  a  los  80  km  de  altura,  la temperatura  disminuye  hasta  alcanzar  los  -140  °C;  a  partir  de  aquí,  la  temperatura  va  ascendiendo en altura al absorber las radiaciones de alta energía, pudiendo alcanzar más de 1000 ºC a unos 600 km de altura; a partir de aquí la baja densidad de gases impide la transmisión del calor y carece de sentido hablar de temperatura (Universidad de Murcia, s.f.). 
<br />
<br />
**III. HUMEDAD ATMOSFÉRICA**
<br />
<br />
El contenido atmosférico de humedad, que comprende el *vapor de agua*, y las *gotitas de agua y los cristales de hielo de las nubes*, queda determinado por la evaporación local, la temperatura del aire y el transporte atmosférico horizontal de humedad. El agua de las nubes, como promedio, es responsable solamente del 4% de la humedad atmosférica (Barry & Chorley, 1999).
<br />
<br />
**IV. VIENTO**
<br />
<br />
Los **vientos locales** son  modulaciones  de  la  circulación  general  de  la  atmosfera,  que  se  desarrollan  a 
menores escalas (decenas de kilómetros). Este tipo de flujo está fuertemente influenciado por la topografía, propiedades de los materiales (como la inercia térmica), uso del suelo, ecosistemas, entre otros. Cuando la pendiente del terreno es muy inclinada, este tipo de circulaciones  predomina  sobre  sistemas  de  mayor  escala.  Este  tipo de  circulación 
prevalece en ambientes donde los sistemas a escala sinóptica son débiles (IDEAM, 2014).
<br />
<br />
En condiciones normales, existe una tendencia generalizada a que las velocidades del viento sean menores al alba, cuando existe una mezcla térmica vertical escasa y, por lo tanto, el aire inferior no participa en la velocidad del aire superior, que se mueve más libremente. Por el contrario, las velocidades de algunos vientos locales son mayores entre las 13:00 y las 14:00 horas, dado que éste es el momento en que el aire tiene la mayor tendencia a moverse verticalmente, debido al calentamiento de la Tierra, lo que le permite, sujeto a efectos de la fricción, unirse al movimiento del aire superior, más libre (Barry & Chorley, 1999).
<br />
<br />
Una de las formas más usuales de representar las direcciones y velocidades de los vientos locales es a través del uso de una *rosa de los vientos* (*Wind Rose*).En ella, se muestran las distribuciones de velocidades del viento y la frecuencia de variación de las direcciones del viento, basándose en observaciones meteorológicas (*Figura 2*).

**V. PRESIÓN ATMOSFÉRICA**
<br />
<br />
La *presión* se puede medir como una fuerza por unidad de superficie. Las unidades utilizadas por los meteorólogos se denominan *milibares* (*mb*), siendo *1 mb* igual a una fuerza de *100 N* actuando sobre *1 metro cuadrado*. La presión media a nivel del mar ($\rho_{0}$) puede estimarse teniendo en cuanta la masa total atmosférica ($M$), la aceleración media debido a la gravedad ($g$) y el radio medio de la Tierra ($R$).
<br />
<br />
El aire es muy compresible, de tal forma que sus capas inferiores son mucho más densas que las de encima. El 50% de la masa total del aire se encuentra por debajo de los 5 km (Barry & Chorley, 1999).

```julia
function mc(x,y)
    n=length(x)
    xs=sum(x)
    ys=sum(y)
    sx=sum(x.^2)
    sxc=(sum(x)).^2
    sxy=sum(x.*y)
    
    delta=(n*sx)-sxc
    A=((sx*ys)-(xs*sxy))/delta
    B=((n*sxy)-(xs*ys))/delta
    E=sqrt(sum((y-A-B*x).^2)/n);
    return A, B, E
end
```
